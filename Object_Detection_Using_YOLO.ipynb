```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow  # For image display in Colab

def detection(image_paths):
    model = YOLO("yolov8l.pt")

    if isinstance(image_paths, str):  # Ensure image_paths is a list
        image_paths = [image_paths]

    for image_path in image_paths:
        frame = cv2.imread(image_path)
        if frame is None:
            print(f"Error loading image: {image_path}")
            continue

        results = model(frame, agnostic_nms=True)

        for result in results:
            boxes = result.boxes.xyxy
            confidences = result.boxes.conf
            class_ids = result.boxes.cls

            for box, confidence, class_id in zip(boxes, confidences, class_ids):
                x1, y1, x2, y2 = map(int, box)
                label = f"{model.model.names[int(class_id)]} {confidence:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Display the resulting frame
        cv2_imshow(frame)  # Use cv2_imshow instead of cv2.imshow in Colab

if __name__ == "__main__":
    # Example usage: provide a list of image paths
    image_paths = ["/content/car.jpeg"]
    detection(image_paths)
```

**VIDEO DETECTION**

```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow  # For image display in Colab

# Define the video file path
input_video_path = "Crossing.mp4"  # Change this to your video file path
output_video_path = "output_video.mp4"

def detection():
    cap = cv2.VideoCapture(input_video_path)
    frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define the codec and create VideoWriter object
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

    model = YOLO("yolov8l.pt")

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, agnostic_nms=True)

        for result in results:
            boxes = result.boxes.xyxy
            confidences = result.boxes.conf
            class_ids = result.boxes.cls

            for box, confidence, class_id in zip(boxes, confidences, class_ids):
                x1, y1, x2, y2 = map(int, box)
                label = f"{model.model.names[int(class_id)]} {confidence:.2f}"
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Write the frame into the file
        out.write(frame)

    cap.release()
    out.release()

if __name__ == "__main__":
    detection()

# Display the video
from IPython.display import HTML
from base64 import b64encode

def display_video(video_path):
    with open(video_path, 'rb') as f:
        video_data = f.read()
    video_base64 = b64encode(video_data).decode('utf-8')
    video_html = f'''
    <video width="640" height="480" controls>
        <source src="data:video/mp4;base64,{video_base64}" type="video/mp4">
    </video>
    '''
    return HTML(video_html)

display_video(output_video_path)
```

```python
import torch
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5s.pt')

cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)


    frame = np.squeeze(results.render())
    cv2_imshow(frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
